```python
import pandas as pd
import numpy as np
import pickle
import textstat
import nltk
import file_load as fl
from sklearn.neural_network import MLPClassifier
nltk.download('vader_lexicon')
import metric_distribution as md
from sklearn.model_selection import train_test_split
import graph_lib as gl
from nltk.corpus import stopwords
from nltk.corpus import words
from nltk.corpus import wordnet

import warnings
warnings.filterwarnings("ignore")
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download("words")
```

# Cleaning Process
### Importing raw file

```python
df = pd.read_json("op_spam_v1.4.json") ##file with all the data gathered togheter in a very nice way, provided by Arpad
df.head()
```

### Getting hotels name

```python
descp = df["description"]
hotel = []
bad_ch = ["_", "1", "2", "3", "4", "5", "6", "7", "8", "9", "0"]
for tag in descp:
    hotel_name = ""
    for ch in tag:
        if ch not in bad_ch:
            hotel_name += ch
    if hotel_name not in hotel:
        hotel.append(hotel_name)
print(hotel)
```

### Tokenizing and removing StopWords and Punctuations

```python
df["token"] = [nltk.word_tokenize(text) for text in df["text"]]

stop_words = set(stopwords.words('english'))
stop_words.remove("no")
stop_words.remove("not")


no_StopWord_token_list = []
for text in df["token"]:
    no_StopWord_token = []
    for word in text:
        if word not in stop_words:
            no_StopWord_token.append(word.lower())            
    no_StopWord_token_list.append(no_StopWord_token)

df["no_StopWord_token"] = no_StopWord_token_list
print("Removing Stop-Words procedure: Done")

#####################################################################
####################################################################

puntk = set([".", ",", ":", "!", "?", '"', ";", "(", ")", "...", "''",])
for text in df["no_StopWord_token"]:
    for word in text:
        if word in puntk:
            text.remove(word)
            
text = df["no_StopWord_token"]
df[["truth", "polarity", "no_StopWord_token"]].head()
```

### Counting and removing errors

```python
hotels_number = hotel
for i in range(0, 1000):
    hotels_number.append(str(i))
    
set_words = set(words.words())
print("Length of the dictionary: {}".format(len(words.words())))

not_error = ["n't", "'s", "'d", "'ll", "'m", "'re", "'ve"]
error = []
error_list = []

for review in text:
    error_count = 0
    error_words = []
    for word in review:
        if word not in set_words and word not in hotels_number:
            if len(wordnet.synsets(word)) == 0 and word not in not_error:
                #not an english word
                error_words.append(word)
                error_count += 1
    error.append(error_count)
    error_list.append(error_words)

df["error_list"] = error_list

#####################################################################
####################################################################

df["number_of_error"] = [len(review) for review in df["error_list"]]

#####################################################################
####################################################################

no_error_token = []
zipped = zip(df["no_StopWord_token"], df["error_list"])

for rev, err in zipped:
    no_err = []
    for w in rev:
        if w not in err:
            no_err.append(w)
    no_error_token.append(no_err)
    
df["no_error_token"] = no_error_token
df.drop(["file", "token"], axis = 1).head()
```

### Importing local features file

```python
metadata_df = fl.pickle_load('raw_data.pickle')
metadata_df = metadata_df.drop('file', axis=1)
metadata_df['truth'] = [int(i) for i in metadata_df['truth']]
metadata_df['polarity'] = [1 if i else 0 for i in metadata_df['polarity']]
```

### Adding error's count

```python
metadata_df["number_of_error"] = df["number_of_error"]
```

### Adding some metadata

```python
mtl = [md.metric_total_length(obj) for obj in metadata_df['text']]
metadata_df['metric_total_length'] = mtl

masl = [md.metric_avg_sentence_length(obj) for obj in metadata_df['text']]
metadata_df['metric_avg_sentence_length'] = masl

mp = [md.metric_polarity(obj) for obj in metadata_df['text']]
metadata_df['metric_polarity'] = mp
metadata_df.drop(["description"], axis=1).head(3)
```

### Adding psycholinguistic features

```python
read_1 = [textstat.flesch_kincaid_grade(obj) for obj in metadata_df['text']]
metadata_df['flesch_kincaid_grade'] = read_1

read_2 = [textstat.coleman_liau_index(obj) for obj in metadata_df['text']]
metadata_df['coleman_liau_index'] = read_2

read_3 = [textstat.automated_readability_index(obj) for obj in metadata_df['text']]
metadata_df['automated_readability_index'] = read_3

metadata_df[['text', 'polarity', 'truth', 'automated_readability_index', 'coleman_liau_index', 'flesch_kincaid_grade']].head(5)
```

### Calculating accuracy 

```python
clf = MLPClassifier(solver='lbfgs', alpha=1/10000,
                    hidden_layer_sizes=(3, 13), random_state=42, learning_rate = 'adaptive')
clf.fit(X_train, y_train)
res = clf.predict(X_test)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_fscore_support
test_score = clf.score(X_test, y_test)
test_F_score = f1_score(y_test, clf.predict(X_test))
print("Accurancy on train: {}\nAccurancy on test: {}"
      .format(clf.score(X_train, y_train), test_score))
```
